```python
from shapely.geometry import Polygon,MultiPolygon
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
p1 =  Polygon([(0,0),(0,1),(1,1),(1,0)])
p2 =  Polygon([(1,1),(1,2),(2,2),(2,1)])
mp = MultiPolygon([p1,p2])
```

```python
fig,ax = plt.subplots()
i = 0
for polygon in  mp:
    i = i+1
    xy= np.array(polygon.exterior.xy).T
    pg = patches.Polygon(xy,closed=True,fill=None,label=f'polygon {i}')
    ax.add_patch(pg)
ax.autoscale_view()
ax.legend()
```
