nuscenes Map Api

`````python
import matplotlib.pyplot as plt
import tqdm
import numpy as np
from nuscenes.map_expansion.map_api import NuScenesMap
from nuscenes.map_expansion import arcline_path_utils
from nuscenes.map_expansion.bitmap import BitMap
`````

`````python
nusc_map = NuScenesMap(dataroot="/myvectormapnet/datasets/nuScenes",map_name="singapore-onenorth")
`````

渲染地图

`````python
fig,ax = nusc_map.render_layers(nusc_map.non_geometric_layers,figsize=1)
`````

激光雷达底图

`````python
bm = BitMap(nusc_map.dataroot,nusc_map.map_name,'basemap')
fig,ax = nusc_map.render_layers(['lane'],figsize=1,bitmap=bm)
`````

渲染一个元素

`````python
fig,ax = nusc_map.render_record('stop_line',nusc_map.stop_line[14]['token'],other_layers=[],bitmap=bm)
`````
